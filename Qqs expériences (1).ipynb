## Effet du paramètre N sur l'indicateur TA

### - N petit (ici, N=5)

```python
import numpy as np
def ex():
    L=[0.5] #p est initialement égal à 0.5
    L1= np.random.uniform(low=0.0, high=1.0, size=4)
    while len(L)<5:
        for i in L1:
            L.append(i)
        return L
    
print(ex())
```

```python
x=np.linspace(0,4,5)
y=ex()
plt.plot(x,y)
plt.xlabel('temps')
plt.ylabel('p')
plt.title('evolution de p en fonction du temps')
plt.show()
```

### - N grand (ici, N=50)

```python

def ex():
    L=[0.5] #p est initialement égal à 0.5
    L1= np.random.uniform(low=0.0, high=1.0, size=49)
    for i in L1:
        L.append(i)
    return L

print(ex())
```

```python
x=np.linspace(0,4,50)
y=ex()
plt.plot(x,y)
plt.xlabel('temps')
plt.ylabel('p')
plt.title('evolution de p en fonction du temps')
plt.show()
```

***Dans tous les cas:
    Pour N grand, le temps d'adoption de la fake news(c'est à dire lorsque p est très proche de 1), est très inférieur à ce même temps lorsque N est petit.  
    Donc, plus il y a des agents dans la matrice, plus les agents seront convaincus rapidement***

## Effet du paramètre news sur TA:

```python
import numpy as np
import matplotlib.pyplot as plt
```

## - news élevée

```python
def icn():
    news=0.9
    L=[0.5]
    N=5
    fp=0.75
    K=np.random.uniform(low=0.0,high=1.0,size=N-1)
    for p in K: #p est une conviction prise àlétaoirement dans une liste de convictions
        i = p   #i est une conviction modifiée par la credibilité de la news
        i = i+fp*(news-i)
        L.append(i)
    return L

print('K=',K)
print(icn())
```

```python
x=np.linspace(0,4,N)
y=icn()
plt.plot(x,y)
plt.xlabel('temps')
plt.ylabel('p')
plt.title('evolution de p en fonction du temps')
plt.show()
```

```python
news = 0.9
print('Crédibilité de l\'info:',news)

def influ_credib_news(news,p,fp):
    #news: crédibilité de la news
    #p: param déterminant le chgt ou non de l'opinion    
    i = p
    i = i+fp*(news-i)
    return i

print(influ_credib_news(news,0.5,0.75))
```
