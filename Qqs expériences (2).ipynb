# Indicateurs

  * TA : Temps d'adoption de la fake news
  * TF : Taux de personnes qui croient à la fin
  * P  : Indicateur sur p (la conviction)

## Effet du paramètre N sur l'indicateur TA

- N est le nombre d'individus

### - N petit (ici, N=5)

```python
import numpy as np
def ex():
    #renvoie une liste de convictions
    L=[0.5] #p est initialement égal à 0.5
    L1= np.random.uniform(low=0.0, high=1.0, size=4)
    while len(L)<5:
        for i in L1:
            L.append(i)
        return L
    
print(ex())
```

```python
x=np.linspace(0,4,5)
y=ex()
plt.plot(x,y)
plt.xlabel('temps')
plt.ylabel('p')
plt.title('evolution de p en fonction du temps')
plt.show()
```

### - N grand (ici, N=50)

```python

def ex2():
    #renvoie une liste de convictions
    L=[0.5] #p est initialement égal à 0.5
    L1= np.random.uniform(low=0.0, high=1.0, size=49)
    for i in L1:
        L.append(i)
    return L

print(ex2())
```

```python
x=np.linspace(0,4,50)
y=ex2()
plt.plot(x,y)
plt.xlabel('temps')
plt.ylabel('p')
plt.title('evolution de p en fonction du temps')
plt.show()
```

***Dans tous les cas:
    Pour N grand, le temps d'adoption de la fake news(c'est à dire lorsque p est très proche de 1), est très inférieur à ce même temps lorsque N est petit.  
    Donc, plus il y a des agents dans la matrice, plus les agents seront convaincus rapidement***

## Effet du paramètre news sur TA:

```python
import numpy as np
import matplotlib.pyplot as plt
```

## - news élevée

```python
news=0.9

def icn():
    #renvoie une liste de convictions
    L=[0.5]
    N=5
    fp=0.75
    K=np.random.uniform(low=0.0,high=1.0,size=N-1)
    for p in K: #p est une conviction prise aléatoirement dans une liste de convictions
        i = p   #i est une conviction modifiée par la credibilité de la news
        i = i+fp*(news-i)
        L.append(i)
    return L

print(icn())
```

```python
x=np.linspace(0,4,5)
y=icn()
plt.plot(x,y)
plt.xlabel('temps')
plt.ylabel('p')
plt.title('evolution de p en fonction du temps')
plt.show()
```

## - news faible

```python
news=0.2

def icn2():
    #renvoie une liste de convictions
    L=[0.5]
    N=5
    fp=0.75
    K=np.random.uniform(low=0.0,high=1.0,size=N-1)
    for p in K: #p est une conviction prise àlétaoirement dans une liste de convictions
        i = p   #i est une conviction modifiée par la credibilité de la news
        i = i+fp*(news-i)
        L.append(i)
    return L


print(icn2())
```

```python
x=np.linspace(0,4,5)
y=icn2()
plt.plot(x,y)
plt.xlabel('temps')
plt.ylabel('p')
plt.title('evolution de p en fonction du temps')
plt.show()
```

***Plus news est grand (i.e. la news est crédible), moins le temps d'adoption de la fake news est long.***

## Effet du paramètre news sur TF

## - news grand

```python
news=0.9

def test():
    #renvoie la liste des convictions sous l'influence de la crédibilité de la news
    Test_reseau={str(k):(np.random.choice([0,1]),np.random.choice([False,True]),np.random.random_sample(),np.random.random_sample()) for k in range(1,11)}
    M=[]
    fp=0.75
    for (y,(a,b,x,c)) in Test_reseau.items():
        c = c+fp*(news-c)
        M.append(c)
    return M

print(test())
```

```python
def tcf():
    #
    R=[]
    for e in test():
        if e>0.8:
            R.append(e)
    return len(R)

print(tcf())
```

## Effet du paramètre news sur P

## - news élevée 

```python
news = 0.9
print('Crédibilité de l\'info:',news)

def influ_credib_news(news,p,fp):
    #news: crédibilité de la news
    #p: param déterminant le chgt ou non de l'opinion    
    i = p
    i = i+fp*(news-i)
    return i

print(influ_credib_news(news,0.5,0.75))
```
